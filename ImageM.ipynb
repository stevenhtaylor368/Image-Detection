```python
import numpy as np  #Nimport Numerical Python 
import cv2          #import OpenCV 2
import os
import glob

def makeCarOut(dir):
    a = []
    output = []
    data_path = os.path.join(dir,'*.jpg') 
    files = glob.glob(data_path) 
    for i in files:
        img = cv2.imread(i)
        a.append(img)   
    for img in a:
        z =  img.copy()
        output.append(z)
    count =0
    for y in output: 
        cv2.imwrite('Images/train/car/C' + str(count) + '.jpg', y)
        count= count +1
    
def makePlaneOut(dir):
    a = []
    output = []
    data_path = os.path.join(dir,'*.jpg') 
    files = glob.glob(data_path) 
    for i in files:
        img = cv2.imread(i)
        a.append(img)   
    for img in a:
        z =  img.copy()
        output.append(z)
    count =0
    for y in output: 
        cv2.imwrite('Images/train/plane/A' + str(count) + '.jpg', y)
        count= count +1

def makeMotorCOut(dir):
    a = []
    output = []
    data_path = os.path.join(dir,'*.jpg') 
    files = glob.glob(data_path) 
    for i in files:
        img = cv2.imread(i)
        a.append(img)   
    for img in a:
        z =  img.copy()
        output.append(z)
    count =0
    for y in output: 
        cv2.imwrite('Images/train/motorcycle/M' + str(count) + '.jpg', y)
        count= count +1
def makeFaceOut(dir):
    a = []
    output = []
    data_path = os.path.join(dir,'*.jpg') 
    files = glob.glob(data_path) 
    for i in files:
        img = cv2.imread(i)
        a.append(img)   
    for img in a:
        z =  img.copy()
        output.append(z)
    count =0
    for y in output: 
        cv2.imwrite('Images/train/face/F' + str(count) + '.jpg', y)
        count= count +1
def makeLeavesOut(dir):
    a = []
    output = []
    data_path = os.path.join(dir,'*.jpg') 
    files = glob.glob(data_path) 
    for i in files:
        img = cv2.imread(i)
        a.append(img)   
    for img in a:
        z =  img.copy()
        output.append(z)
    count =0
    for y in output: 
        cv2.imwrite('Images/train/leaf/L' + str(count) + '.jpg', y)
        count= count +1
def main():
    #cars
    car_dir = "C:\\Users\\Steven\\Pictures\\CSCI412\\Cars" #dir for img
    makeCarOut(car_dir)
    print("cars done")
    plane_dir = "C:\\Users\\Steven\\Pictures\\CSCI412\\Plane" #dir for img
    makePlaneOut(plane_dir)
    print("plane done")
    motor_dir = "C:\\Users\\Steven\\Pictures\\CSCI412\\Motorcycle" #dir for img
    makeMotorCOut(motor_dir)
    print("motor done")
    face_dir = "C:\\Users\\Steven\\Pictures\\CSCI412\\Faces"
    makeFaceOut(face_dir)
    print("face done")
    leaf_dir = "C:\\Users\\Steven\\Pictures\\CSCI412\\Leaves"
    makeLeavesOut(leaf_dir)
    print("leaves done")

main()
```
